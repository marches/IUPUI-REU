```python
import pandas as pd
import numpy as np
from math import *
from random import *
import datetime as datetime
import matplotlib.pyplot as plt
import math
from numpy import unravel_index
from PointProcess import PointProcessTrain
from PointProcess import PointProcessRun
from wasserstein import Cluster
from numpy import array
```

```python
save_loc = 'Trained_Params_70_50.npz'
simulation = PointProcessRun(param_location = save_loc, save_loc = 'PPtesting.npz')
start_time = simulation._LastTime
num = 60
future_intensity, times, incr = simulation.get_future_events(start_time, num, 0)
wasserstein_locs = simulation.locs_for_wasserstein(start_time, num, top_percent=80)
np.amax(future_intensity), future_intensity.shape
```

```python
future_intensity = sum(future_intensity)
np.amax(future_intensity), future_intensity.shape
```

```python

x_labels = []
y_labels = []
for i in range(0, len(future_intensity)):
    x, y = simulation.grid_to_coord(i,i)
    y_labels.append(str(round(y,3)))
    x_labels.append(str(round(x,3)))
y_labels = y_labels[:50]
len(y_labels)
```

```python
int_df = pd.DataFrame(np.transpose(future_intensity), columns = x_labels, index = y_labels)
int_df
```

```python
int_df = int_df.reindex(index=int_df.index[::-1])
```

```python
cluster = Cluster(wasserstein_locs, n = 5)
cluster.randomize_centers()
lam = cluster.learn_lam(n_iter=28, rand_centers=False)
centers= cluster.get_centers()
centers_grid_x = np.zeros(len(centers))
centers_grid_y = np.zeros(len(centers))

centers
```

```python
xsize = simulation._xsize
ysize = simulation._ysize

for i in range(0, len(centers)):
    centers_grid_x[i], centers_grid_y[i] = simulation.coord_to_grid(centers[i][1], centers[i][0])
centers_grid_x, centers_grid_y
```

```python
# add alpha (transparency) to a colormap
import matplotlib.colors 
import matplotlib.cm 

plt.subplots(figsize=(150,100))

# get the map image as an array so we can plot it 
import matplotlib.image as mpimg 
map_img = mpimg.imread('/home/march/IUPUI-REU/0Screenshots/stitched_together_map.png') 

import seaborn as sns; sns.set()
sns.set(font_scale=15)
ax = sns.heatmap(int_df,
            cmap = matplotlib.cm.hot,
            alpha = 0.40, # whole heatmap is translucent
            annot = False,
            zorder = 2,
            xticklabels=10,
            yticklabels=10,
            rasterized=True,
            linewidths=0.0
            )


# heatmap uses pcolormesh instead of imshow, so we can't pass through 
# extent as a kwarg, so we can't mmatch the heatmap to the map. Instead, 
# match the map to the heatmap:

ax.imshow(map_img,
          aspect = ax.get_aspect(),
          extent = ax.get_xlim() + ax.get_ylim(),
          zorder = 1) #put the map under the heatmap

ax.scatter(centers_grid_x,centers_grid_y,marker='.', s = 45000, color=[(25/255,83/255,95/255)], zorder=3, label="Ambulance Placements")
legend = ax.legend(loc='lower left',frameon=True)
legend.get_frame().set_facecolor('w')
from matplotlib.pyplot import show 
sns.set(font_scale=15)

#ax.set_title("Projected Intensities & Ambulance Placements Over a 4 Hour Period")
show()
```
